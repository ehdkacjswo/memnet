```python
def extract_revised_version(input_string):
    keyword = "Rephrased Version: "
    index = input_string.find(keyword)

    if index != -1:
        result = input_string[index + len(keyword):].strip()
        return result
    else:
        return "String not found"
```

```python
def examine_output(output):
  output_tokens = nltk.word_tokenize(output)

  total = len(output_tokens)
  total_common = 0
  problematic = set()
  for token in output_tokens:
    if token in common_1000_wordlist or stemmer.stem(token) in common_1000_wordlist:
      total_common+=1
      continue
    if token.isalpha():
      problematic.add(token)

  return (total_common/total, problematic)
```

```python
# Variables
MAX_ITERATION = 3
MIN_PROPORTION = 0.75
```

```python
def generate_output(input):
  output = llm(input, stream=False)
  #print(output)

  output = extract_revised_version(output)

  proportion, problematic = examine_output(output)

  proportion_mapped_outputs = [(proportion,output)]

  i = 0
  while proportion <MIN_PROPORTION and i < MAX_ITERATION:
    problem_prompt = "I want you to try to change these words to more commonly used words in particular: " + (", ").join(problematic)
    #print("Prompt: " + problem_prompt)

    prompt = f'''Can you please rephrase the following term of service using simpler language that is easy to understand?
    "{output}" {problem_prompt} Give me an output in the form "Rephrased Version: <rephrased term of service>. Your response must have the string 'Rephrased Version:' in it!" \
   "You must have a simplified version that follows the string 'Rephrased Version"
    '''

    tokens = llm.tokenize(prompt)

    output = llm(prompt, stream=False)
    #print(output)

    output = extract_revised_version(output)

    proportion, problematic = examine_output(output)
    proportion_mapped_outputs.append((proportion,output))

    i += 1

  best_output = max(proportion_mapped_outputs, key=lambda x: x[0])[1]

  #print(proportion_mapped_outputs)

  return best_output
```

```python
def final_output(term, hazard_type):
  prompt = "Can you please rephrase the following term of service using simpler language that is easy to understand? \n" + f"{term} \n" +\
   "Give me an output in the form: 'Rephrased Version: <rephrased term of service>'. Your response must have the string 'Rephrased Version:' in it!" \
   "You must have a simplified version that follows the string 'Rephrased Version"

  rephrased = generate_output(prompt)

  print(f"You have requested analysis on the term: {term}")
  print()

  hazard_types = {"LTD": "Liability Exclusions and Limitations", \
                  "CR": "Provider's right to unliterally remove content from the service",\
                  "TER": "Provider's right to unilterally terminate the contract", \
                  "CH": "Provider's right to uniliterally modify the contract and/or the service",\
                  "A": "Arbitration on disputes arising from the contract"}


  if hazard_type in hazard_types.keys():
    hazard_text = hazard_types[hazard_type]
    print(f"The given term may be dangerous because it belongs to type category: {hazard_type} -> {hazard_text}")
    print()
  else:
    print("This term does not fall into any of our hazard categories we have at the moment, but we still encourage you to read through the terms with caution.")

  print(f"To better understand the term, this is a simplified version of it: {rephrased}")
  return
```
